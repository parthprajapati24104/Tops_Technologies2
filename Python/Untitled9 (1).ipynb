```python
# prompt:  chat for runtime and genre top 3 highest  genre by ryntime in bar chart

import matplotlib.pyplot as plt
# Group by genre and calculate the mean runtime
genre_runtime = imdb.groupby('genre')['runtime'].mean().sort_values(ascending=False)

# Get the top 3 genres by runtime
top_3_genres = genre_runtime.head(3)

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_3_genres.index, top_3_genres.values)
plt.xlabel("Genre")
plt.ylabel("Average Runtime")
plt.title("Top 3 Genres by Average Runtime")
plt.show()
```

The bar chart highlights the top 3 genres with the highest average runtime.

These genres typically feature longer films, suggesting they may involve more complex storytelling, in-depth character development, or elaborate narratives that require extended screen time.

```python
# prompt: top 3 kind by highest rating bar chart

import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataframe is named 'imdb'
# If not, replace 'imdb' with your dataframe's name

# Group by 'kind' and calculate the mean rating
kind_rating = imdb.groupby('kind')['rating'].mean().sort_values(ascending=False)

# Get the top 3 kinds by rating
top_3_kinds = kind_rating.head(3)

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_3_kinds.index, top_3_kinds.values)
plt.xlabel("Kind")
plt.ylabel("Average Rating")
plt.title("Top 3 Kinds by Average Rating")
plt.show()
```

The bar chart illustrates the top 3 content types (kinds) with the highest average IMDb ratings.

```python
# prompt: top  country which have highest  runtime bar  chart

import matplotlib.pyplot as plt
# Assuming your dataframe is named 'imdb'
# If not, replace 'imdb' with your dataframe's name

# Group by country and calculate the mean runtime
country_runtime = imdb.groupby('country')['runtime'].mean().sort_values(ascending=False)

# Get the top 10 countries by runtime
top_10_countries = country_runtime.head(10)

# Create the bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10_countries.index, top_10_countries.values)
plt.xlabel("Country")
plt.ylabel("Average Runtime")
plt.title("Top 10 Countries by Average Runtime")
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

The bar chart displays the top 10 countries with the highest average movie runtime.

 This suggests that films from these countries tend to be longer, potentially reflecting regional storytelling traditions, cinematic styles, or audience preferences for more detailed narratives

```python
# prompt: genre vs  counte genre line bar chart

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


genre_country_counts = imdb.groupby(['genre', 'country']).size().reset_index(name='counts')


top_10_combinations = genre_country_counts.nlargest(10, 'counts')

plt.figure(figsize=(12, 6))
sns.barplot(x='genre', y='counts', hue='country', data=top_10_combinations)
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.title("Top 10 Genre-Country Combinations")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The bar chart showcases the top 10 most frequent genre-country combinations in terms of movie production.

 It reveals which genres are most popular in specific countries, highlighting dominant production trends and cultural preferences

```python
# prompt: total show by country bar chart

import matplotlib.pyplot as plt

# Group by country and sum the runtime
country_total_runtime = imdb.groupby('country')['runtime'].sum().sort_values(ascending=False)

# Get the top 10 countries by total runtime
top_10_countries = country_total_runtime.head(10)

# Create the bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10_countries.index, top_10_countries.values)
plt.xlabel("Country")
plt.ylabel("Total Runtime")
plt.title("Top 10 Countries by Total Runtime")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The bar chart highlights the top 10 countries with the highest total runtime across all their films.

These countries typically produce a large volume of content or feature extended-length films, contributing significantly to the global runtime.

```python
# prompt: title by highest vote pie chrt

import matplotlib.pyplot as plt

# Group by 'title' and sum the 'vote'
title_votes = imdb.groupby('title')['vote'].sum().sort_values(ascending=False)

# Get the top 10 titles by vote
top_10_titles = title_votes.head(10)

# Create the pie chart
plt.figure(figsize=(10, 8))
plt.pie(top_10_titles, labels=top_10_titles.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Top 10 Titles by Total Votes')
plt.show()
```

The pie chart illustrates the distribution of total votes across the top 10 titles.

The largest portion is likely attributed to a single title or a few highly popular movies, indicating widespread audience engagement and interest.

 The other titles, while still significant, hold a smaller share, suggesting varying levels of popularity.
